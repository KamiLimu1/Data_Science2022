## Correlation coefficients between numeric features and SalePrice

```python
numeric_df = train_copy.select_dtypes(include=[np.number])
numeric_df.columns
```

```python
correlation = numeric_df.corr()
correlation['SalePrice'].sort_values(ascending = False)
```

### Correlation Visualisation
#### 1. Zoomed Heatmap

```python
# Plot heatmap of the first 11 columns correlating with SalePrice
cols = correlation.nlargest(11, 'SalePrice')['SalePrice'].index
print(cols)

cm = np.corrcoef(train_copy[cols].values.T)
ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='plasma',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values);
```

#### Observations
- `'OverallQual'`, `'GrLivArea'`,`'GarageCars'` and `'GarageArea'` are strongly correlated to `'SalePrice'`.
- `'TotBsmtSF'` and `'1stFlrSF'` are identical. Drop `'1stFlrSF'` in this case as it is less correlated to `'TotBsmtSF'`.
- `'GarageCars'` and  `'GarageArea'` are closely correlated. This is because the number of cars that fit into the garage depend on the garage area. I'll keep `'GarageCars'` since it is strongly correlated to SalePrice than GarageArea 
- `'TotRmsAbvGrd'` and `'GrLivArea'` are also strongly correlated.

### Keep the most correlated variables 

```python
train = train_copy[['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars',
       'TotalBsmtSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt',
       'YearRemodAdd']]
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(14,10))

OverallQual_scatter_plot = pd.concat([train['SalePrice'],train['OverallQual']],axis = 1)
sns.regplot(x='OverallQual',y = 'SalePrice',data = OverallQual_scatter_plot,scatter= True, fit_reg=True, ax=ax1)

TotalBsmtSF_scatter_plot = pd.concat([train['SalePrice'],train['TotalBsmtSF']],axis = 1)
sns.regplot(x='TotalBsmtSF',y = 'SalePrice',data = TotalBsmtSF_scatter_plot,scatter= True, fit_reg=True, ax=ax2)

GrLivArea_scatter_plot = pd.concat([train['SalePrice'],train['GrLivArea']],axis = 1)
sns.regplot(x='GrLivArea',y = 'SalePrice',data = GrLivArea_scatter_plot,scatter= True, fit_reg=True, ax=ax3)

FullBath_scatter_plot = pd.concat([train['SalePrice'],train['FullBath']],axis = 1)
sns.regplot(x='FullBath',y = 'SalePrice',data = FullBath_scatter_plot,scatter= True, fit_reg=True, ax=ax4)

YearBuilt_scatter_plot = pd.concat([train['SalePrice'],train['YearBuilt']],axis = 1)
sns.regplot(x='YearBuilt',y = 'SalePrice',data = YearBuilt_scatter_plot,scatter= True, fit_reg=True)

YearRemodAdd_scatter_plot = pd.concat([train['SalePrice'],train['YearRemodAdd']],axis = 1)
YearRemodAdd_scatter_plot.plot.scatter('YearRemodAdd','SalePrice');
# , ax=ax5
# ,(ax5,ax6
```

### SalePrice vs Overall Quality

```python
saleprice_overall_quality= train.pivot_table(index ='OverallQual',values = 'SalePrice', aggfunc = np.median)
saleprice_overall_quality.plot(kind = 'bar')
plt.xlabel('Overall Quality')
plt.ylabel('Median Sale Price')
plt.show()
```

```python
var = 'OverallQual'
data = pd.concat([train['SalePrice'], train[var]], axis=1)
plt.subplots(figsize=(12, 8))
fig = sns.boxplot(x=var, y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000);
```
